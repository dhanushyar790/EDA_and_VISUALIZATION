```python
#Linegraph--> To track changes, patterns over time(can take one or two features)
plt.plot(df["BloodPressure"],df["Insulin"])
plt.xlabel("BloodPressure")
plt.ylabel("Insulin")
plt.show()
```

```python
#Scatterplot --> To find the correlation between two columns or continuous labels, when you need to understand how two variables relate to each other(Does X affect Y)
plt.scatter(df["Glucose"],df["Insulin"])
plt.xlabel("glucose")
plt.ylabel("Insulin")
plt.show()
```

```python
#Barplot --> used to compare different groups(Which one is biggest,smallest or most popular),used for also categories
plt.bar(df["SkinThickness"],df["BMI"])
```

```python
# Histogram --> show data distribution,data frequency( without bins it become impossible to read)
plt.hist(df["SkinThickness"])
```

```python
#Using bins
plt.hist(df["SkinThickness"],bins=5)
#BINS ==> how many equal sized intervals(Look at my lowest number and my highest number, and slice that distance into exactly X number of equal pieces.)
```

**Use a histogram when you have a large list of numbers and want to answer questions like: "Is my data symmetrical? Where is the middle? Are there any extreme values?"**


**A subplot is a smaller, independent graph that shares the same display window or canvas with other graphs. Instead of opening multiple separate windows or forcing all your data onto a single messy chart, you divide your screen into a grid (like a chessboard or a photo collage) and place one chart into each cell.**

```python
cols=df.columns.tolist()  # List the number of columns in the data
```

```python

plt.figure(figsize=(14, 10))

for i in range(len(cols)):
    plt.subplot(3, 3, i + 1)
    plt.hist(df[cols[i]], bins=15)
    plt.title(cols[i])

# Adjust spacing so titles and axes do not overlap
plt.tight_layout()
plt.show()
```

```python
cols = df.columns.tolist()
plt.figure(figsize=(14, 10))

for i in range(len(cols)):
    plt.subplot(3, 3, i + 1)
    current_col = cols[i]
    if current_col == "Outcome":
        plt.plot(df[current_col])
    elif current_col == "Age":
        plt.boxplot(df[current_col])
    else:
        plt.hist(df[current_col], bins=15)
    plt.title(current_col)

plt.tight_layout()
plt.show()
```
